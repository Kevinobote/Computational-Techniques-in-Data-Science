```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Random walk
import numpy as np
num_step = 1000 # Number of steps
steps = np.random.choice([-1, 1], size = num_step) # Random choice of -1 or 1
position = np.cumsum(steps) # Cumulative sum of steps
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(position, label='1D Random Walk')
plt.axhline(0, color='black', lw=1, ls='--')
plt.title('1D Random Walk')
plt.xlabel('Steps')
plt.ylabel('Positions')
plt.legend()
plt.show()
```

```python
# Compare multiple 1D Random Walks (Comparing 5 walks)
plt.figure(figsize=(10, 6))
for i in range(5):
    steps = np.random.choice([-1, 1], size = num_step) # Random choice of -1 or 1
    position = np.cumsum(steps) # Cumulative sum of steps
    plt.plot(position, alpha =0.75, label=f'Walk {i+1}')
```

```python
# 2D Random Walk
# define x_position and y_position
num_step = 1000
x_steps = np.random.choice([-1, 1], size=num_step)
y_steps = np.random.choice([-1, 1], size=num_step)
x_position = np.cumsum(x_steps) # Cumulative sum of x_steps
y_position = np.cumsum(y_steps) # Cumulative sum of y_steps


plt.figure(figsize=(10, 6))
plt.plot(x_position, y_position, marker='o', ls='-', alpha=0.7)
plt.scatter(0,0, color='red', label='Start', zorder=3)
plt.scatter(x_position[-1], y_position[-1], color='black', label='End', zorder=3)
plt.title('2D Random Walk')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Higher probability for positive direction
prob_right = 0.7 # Probability of moving right
steps = np.random.choice([-1, 1], size=num_step, p=[1-prob_right, prob_right]) # Generate random steps
position = np.cumsum(steps) # Cumulative sum of steps

plt.figure(figsize=(10, 6))
plt.plot(position, label=' Adjusted 1D Random Walk')
plt.axhline(0, color='black', lw=1, ls='--')
plt.title('Adjusted 1D Random Walk')
plt.xlabel('Steps')
plt.ylabel('Positions')
plt.legend()
plt.show()
```
